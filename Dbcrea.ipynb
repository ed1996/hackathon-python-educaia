```python
spec = es_data[spe]
```

```python
for i in range(0, 75):
    a = random.randint(0, 20)
    b = random.randint(0, 20)
    c = random.randint(0, 20)
    x = random.randint(0, 9)
    y = random.randint(0, 9)
    z = random.randint(0, 9)
    if(a<8):
        a += random.randint(0, 7)
    if(a>16):
        a -= random.randint(0, 3)
    if(b<8):
        b += random.randint(0, 7)
    if(b>16):
        b -= random.randint(0, 3)
    if(c<8):
        c += random.randint(0, 7)
    if(c>16):
        c -= random.randint(0, 3)   
        
    if(x == y):
        x = random.randint(0, 9) != [y,z]
    if(x == z):
        x = random.randint(0, 9) != [y,z]
    if(y == z):
        y = random.randint(0, 9) != [x,z]
    spec[spe[x]][i] = a
    spec[spe[y]][i] = b
    spec[spe[z]][i] = c
#spe[9]
spec.head()
```

```python
for i in range(76, 150):
    a = random.randint(0, 20)
    b = random.randint(0, 20)
    
    x = random.randint(0, 9)
    y = random.randint(0, 9)
    if(a<8):
        a += random.randint(0, 7)
    if(a>16):
        a -= random.randint(0, 3)
    if(b<8):
        b += random.randint(0, 7)
    if(b>16):
        b -= random.randint(0, 3)
    
    if(x == y):
        x = random.randint(0, 9) != y  
    spec[spe[x]][i] = a
    spec[spe[y]][i] = b

#spe[9]
spec
```
